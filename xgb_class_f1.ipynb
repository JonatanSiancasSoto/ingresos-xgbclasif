```python
df = pd.read_csv('df_1.data')
```

```python
df.columns = df.columns.str.strip()
```

```python
df
```

```python
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'Marital_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'target']
```

```python
df.columns = columns
```

```python
df
```

```python
df['target'].unique()
```

```python
df['target'] = np.where(df['target']==' <=50K', 0, 1)
```

```python
df['target'].unique()
```

```python
df.hours_per_week.value_counts()
```

```python
df['m-i-40'] = np.where(df['hours_per_week'] <= 40, 1, 0)
```

```python
df.head()
```

```python
df['m-i-40'].value_counts()
```

```python
df['dif-cap'] = df['capital_gain'] - df['capital_loss']
df.head(20)
```

```python
df['usa'] = np.where(df['native_country'] == ' United-States', 1, 0)
```

```python
df['usa'].value_counts()
```

```python
df['native_country'].value_counts()
```

```python
a = df.groupby('workclass')['workclass'].transform('count')
a
```

```python
df['sex'].unique()
```

```python
df['sex'] = np.where(df['sex']==' Female', 0, 1)
df['sex'].value_counts()
```

```python
df.info()
```

```python
cat_col = ['workclass','education','Marital_status','occupation','relationship','race','native_country']
```

```python
for cat in cat_col:
  df[cat] = df.groupby(cat)[cat].transform('count')
```

```python
df.head()
```

```python
df.rename(columns={'education-num': 'education_num', 'm-i-40': 'm_i_40', 'dif-cap': 'dif_cap'}, inplace=True)
df.head()
```

```python
from google.colab import files
df.to_csv('xgb_class.csv', header=True, index=False)
files.download('xgb_class.csv')
```

```python
x = df.loc[:,df.columns != 'target']
```

```python
y = df.loc[:,'target']
```

```python
x.shape, y.shape, df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train, y_test = train_test_split(x, y ,test_size = 0.5, random_state = 12 )
```

```python
x_train.shape, x_test.shape
```

```python
x_valid , x_test , y_valid, y_test = train_test_split(x_test, y_test ,test_size = 0.5, random_state = 12 )
```

```python
x_train.shape, x_test.shape, x_valid.shape
```

```python
import xgboost
```

```python
xgb = xgboost.XGBClassifier()
```

```python
parameters = {'nthreads' : [1],
              'objective': ['binary:logistic'],
              'learning_rate': [0.05,0.1],
              'n_estimator' : [100, 200]}
```

```python
from sklearn.model_selection import GridSearchCV
#from sklearn.model_selection import TimeSeriesSplit
```

```python
fit_params = {
    'early_stopping_rounds': 10,
    'eval_metric': 'logloss',
    'eval_set' : [(x_test,y_test)]
}
```

```python
clf = GridSearchCV(xgb,parameters,cv=3,scoring='accuracy')
```

```python
clf.fit(x_train, y_train, **fit_params)
```

```python
clf.best_estimator_
```

```python
clf.best_score_
```

```python
from sklearn.metrics import accuracy_score
```

```python
best_xgb = clf.best_estimator_
```

```python
best_xgb.predict(x_valid)
```

```python
y_preds = best_xgb.predict(x_valid)
```

```python
comp = pd.DataFrame({'real' : y_valid, 'preds' : y_preds})
comp.head(25)
```

```python
acc = accuracy_score(y_valid, y_preds)
acc
```

```python
from sklearn.metrics import confusion_matrix
confu = confusion_matrix(y_valid,y_preds)
confu
```
